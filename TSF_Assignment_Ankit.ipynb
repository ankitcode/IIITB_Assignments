```python
# Import required packages
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```python
# Import data and convert date to "YYYY-MM-DD" format and set it as index

data = pd.read_csv('dataset_tk.csv', header = 0)
data.rename(columns = {'Unnamed: 0':'Date'}, inplace = True)
data['Date']= pd.to_datetime(data['Date'], format='%d/%m/%Y %H:%M:%S')
data = data.set_index('Date')
data.head(50)
```

```python
# Get data of Karnataka state only

karnataka_data = data["Karnataka"].to_frame()
karnataka_data.head()
#type(karnataka_data)
```

```python
# Weekly resampled data
resampled_karnataka_data = karnataka_data.resample('W').mean()
resampled_karnataka_data.head()
```

```python
# Plot weekly power consumption time series data to check for missing data
resampled_karnataka_data.plot(figsize=(16, 8))
plt.legend(loc='best')
plt.title('Weekly Mean Power Consumption in Karnataka')
plt.show(block=False)
```

```python
# It can be seen from the plot that there are missing values in the data
# Using linear interpolation to fill in the missing values and plot the same

resampled_karnataka_data = resampled_karnataka_data.assign(\
                                                           resampled_karnataka_Linear_Interpolation=\
                                                           resampled_karnataka_data.Karnataka.interpolate(\
                                                                                                           method=\
                                                                                                          'linear'))
resampled_karnataka_data[['resampled_karnataka_Linear_Interpolation']].plot(figsize=(16, 8))
plt.legend(loc='best')
plt.title('Weekly Mean Power Consumption in Karnataka: Linear interpolation')
plt.show(block=False)
```

```python
# Using linear interpolation to impute missing values

resampled_karnataka_data['Karnataka'] = resampled_karnataka_data['resampled_karnataka_Linear_Interpolation']
resampled_karnataka_data.drop(columns=['resampled_karnataka_Linear_Interpolation'],inplace=True)
```

```python
# Checking missing values
resampled_karnataka_data.isnull().sum()
```

```python
resampled_karnataka_data
```

```python
#Split time series data into training and test set

train_len = 80
train = resampled_karnataka_data[0:train_len] # first 80 weeks as training set
test = resampled_karnataka_data[train_len:] # last 21 weeks as out-of-time test set
```

## Naive method

```python
y_hat_naive = test.copy()
y_hat_naive['naive_forecast'] = train['Karnataka'][train_len-1]
```

### Plot train, test and forecast

```python
plt.figure(figsize=(16,8))
plt.plot(train['Karnataka'], label='Train')
plt.plot(test['Karnataka'], label='Test')
plt.plot(y_hat_naive['naive_forecast'], label='Naive forecast')
plt.legend(loc='best')
plt.title('Naive Method')
plt.show()
```

### Calculate MAPE

```python
# Calculate MAPE
mape = np.round(np.mean(np.abs(test['Karnataka']-y_hat_naive['naive_forecast'])/test['Karnataka'])*100,2)

results = pd.DataFrame({'Method':['Naive method'], 'MAPE': [mape]})
results = results[['Method', 'MAPE']]
results
```

## Simple average method

```python
y_hat_avg = test.copy()
y_hat_avg['avg_forecast'] = train['Karnataka'].mean()
```

### Plot train, test and forecast

```python
plt.figure(figsize=(16,8))
plt.plot(train['Karnataka'], label='Train')
plt.plot(test['Karnataka'], label='Test')
plt.plot(y_hat_avg['avg_forecast'], label='Simple average forecast')
plt.legend(loc='best')
plt.title('Simple Average Method')
plt.show()
```

### Calculate MAPE

```python
mape = np.round(np.mean(np.abs(test['Karnataka']-y_hat_avg['avg_forecast'])/test['Karnataka'])*100,2)

tempResults = pd.DataFrame({'Method':['Simple average method'], 'MAPE': [mape] })
results = pd.concat([results, tempResults])
results = results[['Method', 'MAPE']]
results
```

## Simple moving average method

```python
y_hat_sma = resampled_karnataka_data.copy()
ma_window = 2
y_hat_sma['sma_forecast'] = resampled_karnataka_data['Karnataka'].rolling(ma_window).mean()
y_hat_sma['sma_forecast'][train_len:] = y_hat_sma['sma_forecast'][train_len-1]
```

### Plot train, test and forecast

```python
plt.figure(figsize=(16, 8))
plt.plot(train['Karnataka'], label='Train')
plt.plot(test['Karnataka'], label='Test')
plt.plot(y_hat_sma['sma_forecast'], label='Simple moving average forecast')
plt.legend(loc='best')
plt.title('Simple Moving Average Method')
plt.show()
```

### Calculate MAPE

```python
mape = np.round(np.mean(np.abs(test['Karnataka']-y_hat_sma['sma_forecast'][train_len:])/test['Karnataka'])*100,2)

tempResults = pd.DataFrame({'Method':['Simple moving average forecast'], 'MAPE': [mape] })
results = pd.concat([results, tempResults])
results = results[['Method', 'MAPE']]
results
```

## Simple exponential smoothing

# Auto Regressive methods

### Kwiatkowski-Phillips-Schmidt-Shin (KPSS) test

```python
from statsmodels.tsa.stattools import kpss
kpss_test = kpss(resampled_karnataka_data['Karnataka'])

print('KPSS Statistic: %f' % kpss_test[0])
print('Critical Values @ 0.05: %.2f' % kpss_test[3]['5%'])
print('p-value: %f' % kpss_test[1])

# Null Hypothesis (H0): The series is stationary
# pâˆ’value > 0.05
# Alternate Hypothesis (H1): The series is not stationary
# pâˆ’value <= 0.05
```

```python
# p-value: 0.100000 is greater than 0.05, therefore, the series is stationary
```

### Augmented Dickey-Fuller (ADF) test

```python
from statsmodels.tsa.stattools import adfuller
adf_test = adfuller(resampled_karnataka_data['Karnataka'])

print('ADF Statistic: %f' % adf_test[0])
print('Critical Values @ 0.05: %.2f' % adf_test[4]['5%'])
print('p-value: %f' % adf_test[1])

# Null Hypothesis (H0): The series is not stationary
# pâˆ’value > 0.05
# Alternate Hypothesis (H1): The series is stationary
# pâˆ’value <= 0.05
```

```python
# p-value: 0.000000 is less than 0.05, therefore, the series is stationary
```

## Auto regression method (AR)

```python
from statsmodels.tsa.arima.model import ARIMA
model = ARIMA(train, order=(39, 0, 0)) # p - 39
model_fit = model.fit()
print(model_fit.params)
```

```python
y_hat_ar = test.copy()
y_hat_ar['ar_forecast'] = model_fit.predict(test.index.min(), test.index.max())
```

### Plot train, test and forecast

```python
plt.figure(figsize=(16,8))
plt.plot(train['Karnataka'], label='Train')
plt.plot(test['Karnataka'], label='Test')
plt.plot(y_hat_ar['ar_forecast'][test.index.min():], label='Auto regression forecast')
plt.legend(loc='best')
plt.title('Auto Regression Method')
plt.show()
```

### Calculate MAPE

```python
mape = np.round(np.mean(np.abs(test['Karnataka']-y_hat_ar['ar_forecast'][test.index.min():])/test['Karnataka'])*100,2)

tempResults = pd.DataFrame({'Method':['Autoregressive (AR) method'], 'MAPE': [mape] })
results = pd.concat([results, tempResults])
results = results[['Method', 'MAPE']]
results
```
